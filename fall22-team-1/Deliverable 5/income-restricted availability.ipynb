```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_arr = []
report_names = ['2018.csv', '2020.csv', '2021.csv']
years = []

for i in range(3):
    df = pd.read_csv(report_names[i])
    df = df.dropna(subset=['Zip Code', 'Total Income-Restricted'])
    df['Zip Code'] = df['Zip Code'].astype(int)
    df['Total Income-Restricted'] = df['Total Income-Restricted'].astype(int)
    zipcodes = df['Zip Code'].unique()
    zipcodes.sort()
    total_counts = []
    for j in range(len(zipcodes)):
        total_counts.append(df.groupby('Zip Code')['Total Income-Restricted'].sum()[zipcodes[j]])
    year = np.array([zipcodes, total_counts])
    years.append(year)
years = np.array(years)
print(years)
#print(years.shape)

# Graph for all zip codes
#plt.rcParams["figure.figsize"] = (15,15)
#for i in range(years.shape[2]):
#    plt.plot([2018, 2020, 2021], years[:,1,i], label='Zip Code: '+str(years[0,0,i]))
for i in range(22,24):
    plt.plot([2018, 2020, 2021], years[:,1,i], label='Zip Code: '+str(years[0,0,i]))
plt.legend()
plt.show()
# Graph for Brighton and Allston
# Brighton has 02134 and 02135
```
