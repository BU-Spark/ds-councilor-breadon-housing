```python
import folium
from folium.plugins import HeatMap

default_loc = [42.3601, -71.0589]
df['lat_long'] = list(zip(df.latitude, df.longitude))
m = folium.Map(location=default_loc, zoom_start=10)

heatmap_enf = pd.DataFrame()
heatmap_enf["lat"] = pd.Series(df["lat_long"].value_counts().keys()).str[0]
heatmap_enf["long"] = pd.Series(df["lat_long"].value_counts().keys()).str[1]


#heatmap_df["avg_price"] = pd.Series(df.groupby('lat_long')['price'].mean().values)
```

```python
one_hot = pd.get_dummies(df['violation_type'])
# Drop column B as it is now encoded
df = df.drop('violation_type',axis = 1)
# Join the encoded df
df = df.join(one_hot)
```

```python
df.head()
```

```python
heatmap_enf["violations"] = pd.Series(df.groupby('lat_long')['Enforcement Violations'].mean().values)
```

```python
HeatMap(heatmap_enf, min_opacity = 0.4, blur = 18
       ).add_to(folium.FeatureGroup(name='Heat Map').add_to(m))
folium.LayerControl().add_to(m)
```

```python
m.save('enforcement_violations.html')
```
