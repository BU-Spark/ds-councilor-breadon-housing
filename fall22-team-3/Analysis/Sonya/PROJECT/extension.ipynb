```python
import warnings
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns

#ignore by message
warnings.filterwarnings("ignore")

import pandas as pd

df = pd.read_csv("Violations.2010_2021.Neighborhood.csv", low_memory=False)
print(df.head())

df_census = pd.read_csv("2010_census_tract.csv", low_memory=False)
df_census.head()
```
