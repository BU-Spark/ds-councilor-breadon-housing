```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 

df = pd.read_csv("/Users/sv/Desktop/off_campus_2019_2020.csv", encoding_errors= 'replace')
# df
colleges = df['College/University'].to_numpy()
ug = df['UG'].to_numpy()
g = df['G'].to_numpy()
plt.bar(colleges,ug,color='red')
plt.bar(colleges,g,bottom=ug,color='blue')
plt.xticks(rotation=90)
plt.ylabel('Number of Off-Campus Students')
plt.xlabel("Undergraduate and Graduate Students by University")
plt.title('Off-Campus Student Overview 2019-2020')
colors = {'Undergraduate Students':'red', 'Graduate Students':'blue'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
plt.show()








```
